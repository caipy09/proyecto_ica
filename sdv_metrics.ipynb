Evaluating Column Pair Trends: La evaluación de tendencias de pares de columnas implica evaluar si los datos sintéticos capturan tendencias entre pares de columnas. La tendencia entre dos columnas describe cómo varían entre sí, como su correlación. Una puntuación más alta indica que las tendencias son más parecidas.

Evaluating Column Shapes: Los datos sintéticos capturan la forma de cada columna comparando la distribución general de los datos reales y sintéticos. Cuanto mayor sea la puntuación, más similares serán las distribuciones.
