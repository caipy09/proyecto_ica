```python
import numpy as np

def calcular_pesos_y_consistencia(matrix):
    n = matrix.shape[0]

    # Paso 1: Normalización por columnas
    col_sum = matrix.sum(axis=0)
    normalized_matrix = matrix / col_sum

    # Paso 2: Promedio de filas → pesos aproximados
    weights = normalized_matrix.mean(axis=1)

    # Paso 3: Autovalor máximo λ_max aproximado
    # Cuando A⋅w=n⋅w si es perfectamente consistente (n es autovalor y w autovector de pesos)
    # caso contrario λ_max.w
    
    Aw = matrix @ weights
    lambda_max = np.sum(Aw / weights) / n

    # Paso 4: Índice de consistencia CI
    CI = (lambda_max - n) / (n - 1)

    # Paso 5: Índice Aleatorio RI según Saaty
    RI_dict = {1: 0.00, 2: 0.00, 3: 0.58, 4: 0.90, 5: 1.12,
               6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45, 10: 1.49}
    RI = RI_dict.get(n, None)

    if RI is not None and RI != 0:
        CR = CI / RI
    else:
        CR = 0

    return weights, lambda_max, CI, CR

# Aplicando a la matriz de comparación `comparison_matrix` obtenemos:

pesos, lambda_max, CI, CR = calcular_pesos_y_consistencia(comparison_matrix)

# Mostrar resultados
for i, var in enumerate(variables):
    print(f"{var:30} Peso: {pesos[i]:.4f}")

print(f"\nλ_max = {lambda_max:.4f}")
print(f"CI = {CI:.4f}")
print(f"CR = {CR:.4f} → {'Consistente' if CR < 0.1 else 'Inconsistente'}")
```
